### **2. Univariate Analysis**

#### **2.1. Histogram for the Distribution of Income**

```python
plt.figure(figsize = (10,6))
sns.histplot(df['income'], bins = 50, kde = True, 
             color = 'skyblue', edgecolor = 'black')
plt.title('Distribution of Income', 
          fontsize = 14, fontweight = 'bold')
plt.xlabel('Income ($)')
plt.ylabel('Frequency')
plt.axvline(df['income'].median(), color = 'red',
            linestyle = '--', 
            label = f'Median: ${df['income'].median():,.0f}')
plt.show()
```

#### **2.2. Income Distribution from a Boxplot**

```python
plt.figure(figsize = (10, 6))
sns.boxplot(y = df['income'], color = 'steelblue')
plt.title("Income Distribution Using Boxplot", fontsize = 14, 
          fontweight = 'bold')
plt.ylabel('Income ($)')
plt.axhline(df['income'].mean(), color = 'red',
            linestyle = '-', alpha = 0.8, label = 'Mean')
plt.axhline(df['income'].median(), color = 'black',
            linestyle = '--', alpha = 0.8, label = 'Median')
plt.legend()
plt.tight_layout()
plt.show()
```

#### **2.3. Age Distribution Using Histogram and Violin Plots**

```python
fig, axes = plt.subplots(1, 2, figsize = (14, 5))

sns.histplot(df['age'], bins = 20, kde = True, ax=axes[0],
             color = 'steelblue', edgecolor = 'black')
axes[0].set_title('Age Distribution using Histogram', fontweight = 'bold')
axes[0].axvline(df['age'].mean(), color = 'red', linestyle = '--')

sns.violinplot(y = df['age'], ax = axes[1], color = 'lightgreen')
axes[1].set_title("Age Distribution using Violin Plot", fontweight = 'bold')
axes[1].axhline(df['age'].mean(), color = 'red', linestyle = '--')

plt.tight_layout()
plt.show()
```

#### **2.4. Joint Distribution Plot - Height and Weight**

```python
joint_plot = sns.jointplot(data = df, x = 'height', y = 'weight',
                           kind = 'scatter', alpha = 0.8, color = 'purple',
                           marginal_kws = dict(bins = 50, fill = True))
joint_plot.fig.suptitle('A joint Plot of Weight and Height Distribution',
                        fontsize = 14, fontweight = 'bold')
plt.tight_layout()
plt.show()

corr_val = df['height'].corr(df['weight'])
print(f"Height and Weight Correlation: {corr_val:.3f}")
```

### **3. Categorical Variables Analysis**
* Bar Chart - Gender and Educational Level Distribution
* Pie and Count Plot (Bar Chart) showing the Distribution of Marital Status

#### **3.1. Bar Plot for Gender Distribution**

```python
plt.figure(figsize = (10, 6))
ax = sns.countplot(data = df, x = 'sex', palette = 'Set2',
              edgecolor = 'black', linewidth = 1.5)
plt.title("Gender Distribution", fontsize = 14, fontweight = 'bold')
plt.xlabel("Gender")
plt.ylabel("Frequency")

total = len(df)
for p in ax.patches:
    percentage = f'{100*p.get_height()/total:.1f}%'
    x = p.get_x() + p.get_width()/2
    y = p.get_height() + 10
    ax. annotate (percentage, (x, y), ha ='center', fontweight = 'bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
ax = sns.countplot(data = df, x = 'degree', palette = 'Set2',
              edgecolor = 'black', linewidth = 1.5)
plt.title("Educational Level Distribution", fontsize = 14, fontweight = 'bold')
plt.xlabel("Educational Level")
plt.ylabel("Frequency")

total = len(df)
for p in ax.patches:
    percentage = f'{100*p.get_height()/total:.1f}%'
    x = p.get_x() + p.get_width()/2
    y = p.get_height() + 10
    ax. annotate (percentage, (x, y), ha ='center', fontweight = 'bold')

plt.tight_layout()
plt.show()
```

#### **Horizontal Bar Chart for Educational Level**

```python
plt.figure(figsize = (10, 6))
degree_counts = df['degree'].value_counts()
ax = sns.barplot(x = degree_counts.values, y = degree_counts.index,
                 palette = 'viridis', edgecolor = 'black')
plt.title('Education Level Distribution', fontsize = 14, fontweight = 'bold')
plt.xlabel('Frequency')
plt.ylabel("Educational Levels")

for i, v in enumerate(degree_counts.values):
    ax.text(v+10, i, str(v), va = "center", fontweight = 'bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (14, 6))

marital_counts = df['marital'].value_counts()
color = plt.cm.Set3(np.linspace(0, 1, len(marital_counts)))

sns.pie(marital_counts, labels = marital_counts.index,
            autopct = '%1.1f%%', colors = color, startangle = 90,
            explode = [0.05]* len(marital_counts))
plt.title('Marital Status - For the Pie Chart', fontweight = 'bold')
plt.tight_layout()
plt.show()
```

#### **Pie Chart and a Countplot for Marital Status**

```python
fig, axes = plt.subplots(1, 2, figsize = (14, 6))

marital_counts = df['marital'].value_counts()
color = plt.cm.Set3(np.linspace(0, 1, len(marital_counts)))

axes[0].pie(marital_counts, labels = marital_counts.index,
            autopct = '%1.1f%%', colors = color, startangle = 90,
            explode = [0.05]* len(marital_counts))
axes[0].set_title('Marital Status - For the Pie Chart', fontweight = 'bold')

sns.countplot(data = df, x = "marital", ax = axes[1],
    palette = 'pastel', edgecolor = 'black')
axes[1].set_title('Marital Status - For the Bar Chart', fontweight = 'bold')
axes[1].tick_params(axis = 'x', rotation = 45)

plt.tight_layout()
plt.show()

```
