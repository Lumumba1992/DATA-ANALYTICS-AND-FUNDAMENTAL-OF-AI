### 2. Basic Data Types
Python automatically infers types. Core types: `int`, `float`, `str`, `bool`, `NoneType`.

* Numeric Types

```python
# Numeeric Types
int_val = 42
float_val = 3.14159


# String Type
str_val = "Data Analytics"

# Boolean & None
bool_val = True
missing_val = None

print(f"Types: {type(int_val)}, {type(float_val)}, {type(str_val)}, {type(bool_val)}, {type(missing_val)}")
```

### 3. Type Checking & Conversion
Use `type()` to check, and built-in functions for explicit casting.

```python
x = "100"
y = int(x)          # String → Integer
z = float(y)        # Integer → Float
w = str(z)          # Float → String

print(f"Original: '{x}' → Int: {y} → Float: {z} → String: '{w}'")
print(f"Absolute: {abs(-42)} | Rounded: {round(3.14159, 2)} | Max: {max([3,7,2])}")
```

### 4. Mathematical & Comparison Operators

```python
a, b = 15, 4

print("🔢 Arithmetic:")
print(f"Add: {a+b} | Sub: {a-b} | Mul: {a*b}")
print(f"Div: {a/b} | Floor: {a//b} | Mod: {a%b} | Pow: {a**b}")

print("\n🔍 Comparison:")
print(f"a > b: {a>b} | a == b: {a==b} | a != b: {a!=b}")
print(f"a >= 10: {a>=10} | b <= 5: {b<=5}")
```

This code above is a clear demonstration of **Python operators** using two integer variables. You’ve defined $a$ as 15 and $b$ as 4, and the script shows how Python processes mathematical calculations and logic.

### 🔢 Arithmetic Operators
This section performs calculations on the two numbers:
* **Standard Operators:** You have addition (+), subtraction (-), and multiplication (*).
* **Division Types:** `/` (True Division) returns a float (**3.75**).
    * `//` (Floor Division) chops off the decimal, returning the largest whole integer (**3**).
* **Modulo (`%`):** This returns the **remainder** of the division. Since 4 goes into 15 three times with 3 left over, the result is **3**.
* **Exponentiation (`**`):** This raises $a$ to the power of $b$ ($15^4$), resulting in **50,625**.



---

### 🔍 Comparison Operators
This section evaluates the relationship between values and returns a **Boolean** (`True` or `False`):
* **Equality/Inequality:** `==` checks if values are identical, while `!=` checks if they are different.
* **Magnitude:** `>` and `>=` compare which number is larger or if it meets a specific threshold (like checking if $a$ is at least 10).

> **Pro Tip:** In Python, the modulo operator (`%`) is incredibly useful for checking if a number is even or odd (e.g., `x % 2 == 0`).

### 5. Logical Operators & Control Flow

```python
score = 85

# Conditional Logic
if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
else:
    grade = "C"
print(f"Score: {score} -> Grade: {grade}")
```

```python
# Loops
print("\n🔁 For Loop:")
for i in range(3):
    print(f"Iteration {i+1}")
```

### 6. Collections: Lists & Tuples
Lists are mutable. Tuples are immutable. Both are foundational for data structures.

1. A List is like a whiteboard. You can write things down, erase them, change the order, or add new notes whenever you like.

* Definition: You can change, add, or remove elements after the list has been defined.
* Syntax: Defined using square brackets [].
* Analogy: A shopping list. If you forget milk, you just write it at the bottom.

2. Tuples: The "Immutable" Collection
A Tuple is like a stone carving. Once the information is etched in, it stays that way forever. If you want to change it, you have to throw the whole stone away and carve a brand-new one.

* Definition: Once a tuple is created, its contents cannot be changed, reordered, or deleted.
* Syntax: Defined using parentheses ().
* Analogy: The coordinates of a city (Latitude, Longitude).

If you change the numbers, you aren't "editing" the city; you're talking about a completely different location.

```python
data_list = [10, 20, 30, 40]
data_list.append(50)  # Mutable
print(f"List: {data_list}")

coord_tuple = (5.0, 3.2)
# coord_tuple[0] = 6.0  # ❌ This would raise a TypeError
print(f"Tuple: {coord_tuple}")
```

```python
# Slicing
print(f"First two list items: {data_list[:2]}")
print(f"Last tuple item: {coord_tuple[-1]}")
```

### 7. Introduction to NumPy (Numerical Python)
`ndarray` provides fast, memory-efficient arrays and vectorized operations.

```python
import numpy as np

# Creating Arrays
arr_1d = np.array([1, 2, 3, 4])
arr_2d = np.array([[1, 2], [3, 4]])
zeros_mat = np.zeros((2, 3))
ones_vec = np.ones(4)

print("1D Array:", arr_1d)
print("2D Matrix:\n", arr_2d)
print("Shape:", arr_2d.shape)
print("zeros_mat:",zeros_mat)
print("One Vector:", ones_vec)
```

### 8. Mathematical & Matrix Operations

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("➕ Element-wise Addition:\n", A + B)
print("✖️ Element-wise Multiplication:\n", A * B)
print("➗ Scalar Division:\n", A / 2)
```

### 9. Matrix Multiplication & Transpose
Essential for linear algebra, ML, and data transformations.

```python
# Matrix Multiplication (Dot Product)
print("🔗 Matrix Multiplication (A @ B):\n", A @ B)

# Transpose (Rows ↔ Columns)
print("↔️ Transpose of A:\n", A.T)

# Determinant & Inverse (Advanced Math)
print("📐 Determinant:", np.linalg.det(A))
print("🔄 Inverse:\n", np.linalg.inv(A))
```

### 10. Introduction to Pandas
Built on NumPy. Core objects: `Series` (1D) and `DataFrame` (2D table).

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [24, 30, 22, 28],
    "Score": [88.5, 92.0, 79.3, 95.1],
    "Department": ["CS", "Math", "Physics", "CS"]
}

df = pd.DataFrame(data)
print("✅ Created DataFrame:")
#print(df)
df
```

### 11. DataFrame Indexing & Selection

```python
# Single Column (Returns Series)
print("📊 Names Column:\n", df["Name"])
```

```python
# Multiple Columns
print("\n📑 Subset (Name, Score):\n", df[["Name", "Score"]])
```

```python
# Position-based (iloc) & Label-based (loc)
print("\n📍 First Row (iloc[0]):\n", df.iloc[0])
```

```python
print("\n🔎 Filter: Age > 25:\n", df.loc[df["Age"] > 25])
```

### 12. Importing CSV Files
`pd.read_csv()` is the standard for tabular data ingestion.

```python
# Save current DataFrame as CSV for demonstration
df.to_csv("students_sample.csv", index=False)
print("📁 Created 'students_sample.csv'")
```

```python
# Read it back
df_loaded = pd.read_csv("students_sample.csv")
print("\n📥 Loaded from CSV:")
print(df_loaded.head())
```
