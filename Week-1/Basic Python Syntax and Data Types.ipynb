### **Mathematical Operations in Python**

```python
x, y = 18, 5
print('='*60)
print("ARITHMENTIC OPERATIONS")
print('='*60)
print(f"Addition: {x +y} | Subtraction: {x-y} | Multiplication : {x*y}")
```

#### **Division, Modulus, Floor and Power**

```python
print('='*60)
print("ADVANCED MATHEMATICAL OPERATION")
print('='*60)
print(f"Division: {x/y} | Floor: {x//y} | Modulus: {x%y} | Power: {x**2}")
```

### **Logical Operators and Work Flows**

```python
score = 39
score
```

### **Conditional Logic**

```python
if score >= 70:
    grade = "A"
elif score >= 60:
    grade = "B"
elif score >= 50:
    grade = "C"
elif score >= 40:
    grade = "D"
else: 
    grade = "Fail"
```

```python
print(f"Score: {score} -> Grade: {grade}")
```
